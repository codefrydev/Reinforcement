## Exercise 1 — Assignment Step

For each point, find the index of the nearest centroid.

```python
points    = np.array([[1,1],[1,2],[2,1],[8,8],[8,9],[9,8]], dtype=float)
centroids = np.array([[1,1],[8,8],[5,5]], dtype=float)

# TODO: for each point, compute distance to each centroid, argmin
assignments = []
for p in points:
    dists   = None  # np.linalg.norm(centroids - p, axis=1)
    cluster = None  # np.argmin(dists)
    assignments.append(cluster)

print(f'Assignments: {assignments}')
# expected: [0, 0, 0, 1, 1, 1]  (first 3 → centroid 0, last 3 → centroid 1)
```

## Exercise 2 — Update Step

Compute new centroids as mean of assigned points.

```python
assignments = np.array([0, 0, 0, 1, 1, 1])
K = 3
new_centroids = np.zeros((K, 2))

for k in range(K):
    mask = assignments == k
    if mask.sum() > 0:
        # TODO: mean of assigned points
        new_centroids[k] = None  # points[mask].mean(axis=0)
    else:
        new_centroids[k] = centroids[k]  # keep if empty cluster

print('New centroids:')
for k, c in enumerate(new_centroids):
    print(f'  Cluster {k}: {c}')
# expected: cluster 0 ≈ [1.33, 1.33], cluster 1 ≈ [8.33, 8.33], cluster 2 unchanged
```

## Exercise 3 — Full K-Means + Visualization

Run 3 iterations and plot the final clusters.

```python
points    = np.array([[1,1],[1,2],[2,1],[8,8],[8,9],[9,8]], dtype=float)
centroids = np.array([[1,1],[8,8],[5,5]], dtype=float)
K = 3

for iteration in range(3):
    dists = np.linalg.norm(points[:, None, :] - centroids[None, :, :], axis=2)
    assignments = np.argmin(dists, axis=1)
    for k in range(K):
        mask = assignments == k
        if mask.sum() > 0:
            centroids[k] = points[mask].mean(axis=0)

colors = ['blue', 'red', 'green']
for k in range(K):
    mask = assignments == k
    plt.scatter(points[mask, 0], points[mask, 1], c=colors[k], label=f'Cluster {k}', s=100)
    plt.scatter(*centroids[k], c=colors[k], marker='x', s=200, linewidths=3)

plt.title('K-Means Clusters (x = centroid)')
plt.legend()
plt.show()
print(f'Final assignments: {assignments}')
```
