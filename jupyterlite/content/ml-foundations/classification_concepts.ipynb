# Classification Concepts

In this notebook you will implement the sigmoid function, apply decision thresholds, and visualise how the sigmoid maps any real number to a probability between 0 and 1.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Exercise 1: Implement and verify sigmoid

Implement σ(z) = 1 / (1 + exp(-z)) and verify key values.

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Key values
print(f'σ(0)  = {sigmoid(0):.4f}    (expected: 0.5000)')    # maximum uncertainty
print(f'σ(1)  = {sigmoid(1):.4f}    (expected: 0.7311)')
print(f'σ(-1) = {sigmoid(-1):.4f}    (expected: 0.2689)')
print(f'σ(10) = {sigmoid(10):.6f}  (expected: ~1.0)')
print(f'σ(-10)= {sigmoid(-10):.6f}  (expected: ~0.0)')
```

## Exercise 2: Classify with threshold

Given z-scores, compute probabilities with sigmoid, then classify with threshold p > 0.5.

```python
z_scores = np.array([-3.0, -1.0, -0.5, 0.5, 1.0, 3.0])

probs       = sigmoid(z_scores)
predictions = (probs > 0.5).astype(int)

print('z_scores:   ', z_scores)
print('probs:      ', probs.round(3))
print('predictions:', predictions)
print()
print('Decision boundary is at z=0 (prob=0.5)')
```

## Exercise 3: Plot the sigmoid curve

Visualise the sigmoid function and mark the decision boundary at z=0.

```python
z = np.linspace(-6, 6, 200)
p = sigmoid(z)

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(z, p, color='#4C72B0', linewidth=2.5, label='σ(z)')
ax.axhline(y=0.5, color='red', linestyle='--', alpha=0.7, label='threshold = 0.5')
ax.axvline(x=0.0, color='gray', linestyle=':', alpha=0.5, label='decision boundary z=0')
ax.fill_between(z, 0, p, where=(z > 0), alpha=0.1, color='green', label='predict class 1')
ax.fill_between(z, 0, p, where=(z < 0), alpha=0.1, color='red',   label='predict class 0')
ax.set_title('Sigmoid function σ(z)')
ax.set_xlabel('z (linear output)')
ax.set_ylabel('σ(z) = probability')
ax.legend(loc='upper left', fontsize=9)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## Exercise 4: Effect of different thresholds

Compare predictions using threshold 0.5 vs 0.7.

```python
z_scores = np.array([-2., -1., 0., 0.5, 1., 2.])
probs = sigmoid(z_scores)

pred_05 = (probs > 0.5).astype(int)
pred_07 = (probs > 0.7).astype(int)

df = pd.DataFrame({
    'z':          z_scores,
    'prob':       probs.round(3),
    'thresh_0.5': pred_05,
    'thresh_0.7': pred_07,
})
print(df.to_string(index=False))
print('\nHigher threshold (0.7) is more conservative — fewer positives.')
```
