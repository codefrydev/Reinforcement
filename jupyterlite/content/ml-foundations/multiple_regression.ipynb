# Multiple Regression

In this notebook you will implement vectorised multiple regression using NumPy matrix operations: ŷ = Xw + b, MSE loss, and matrix gradient descent.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Exercise 1: Matrix predictions and MSE

Use matrix operations (X @ w + b) to compute predictions and MSE for a multi-feature dataset.

```python
# 5 samples, 3 features: [size_sqm, rooms, age_years]
X = np.array([
    [80,  3, 10],
    [120, 4,  5],
    [60,  2, 20],
    [150, 5,  2],
    [100, 3,  8],
], dtype=float)

y = np.array([250000, 380000, 180000, 480000, 310000], dtype=float)

w = np.array([2000.0, 15000.0, -500.0])   # initial weights
b = 50000.0

# Forward pass
y_hat = X @ w + b
mse   = np.mean((y - y_hat) ** 2)

print('y_hat:', y_hat)
print('y:    ', y)
print(f'MSE:  {mse:.2f}')
print(f'RMSE: {np.sqrt(mse):.2f}')
```

## Exercise 2: Compute gradient and take one step

```python
n      = len(y)
lr     = 1e-7

residuals = y - y_hat
grad_w    = (-2 / n) * (X.T @ residuals)    # shape (3,)
grad_b    = (-2 / n) * np.sum(residuals)

w_new = w - lr * grad_w
b_new = b - lr * grad_b

print('grad_w:', grad_w)
print('w_new: ', w_new)
print(f'b_new:  {b_new:.2f}')
```

## Exercise 3: Full gradient descent with loss plot

Run 5000 steps of gradient descent and plot MSE over time.

```python
w = np.zeros(3)
b = 0.0
lr = 1e-8
losses = []

for step in range(5000):
    y_hat = X @ w + b
    mse   = np.mean((y - y_hat) ** 2)
    losses.append(mse)
    residuals = y - y_hat
    w -= lr * (-2/n) * (X.T @ residuals)
    b -= lr * (-2/n) * np.sum(residuals)

print(f'Final w: {w}')
print(f'Final b: {b:.2f}')
print(f'Final MSE: {losses[-1]:.2f}')

plt.figure(figsize=(7, 4))
plt.plot(losses)
plt.title('MSE loss during multiple regression training')
plt.xlabel('Step')
plt.ylabel('MSE')
plt.yscale('log')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## Exercise 4: Compare to scikit-learn

Compare your gradient descent result to the exact solution from sklearn.

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X, y)

print('sklearn coef_:      ', model.coef_)
print('sklearn intercept_: ', model.intercept_)
print()
print('our w:              ', w)
print('our b:              ', b)
```
