# Linear Regression

In this notebook you will implement linear regression from scratch: compute MSE loss, derive the gradient, and take gradient descent steps in NumPy.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Exercise 1: Compute MSE and one gradient step

Given house size and price data, compute predictions, MSE loss, gradient w.r.t. w, and take one update step.

```python
sizes  = np.array([50, 75, 100, 125, 150], dtype=float)
prices = np.array([150000, 210000, 280000, 340000, 400000], dtype=float)

w = 2500.0
b = 25000.0
lr = 0.0001

# Predictions
y_hat = w * sizes + b

# MSE
mse = np.mean((prices - y_hat) ** 2)

# Gradient
n      = len(sizes)
grad_w = (-2 / n) * np.sum(sizes * (prices - y_hat))
grad_b = (-2 / n) * np.sum(prices - y_hat)

# One step
w_new = w - lr * grad_w
b_new = b - lr * grad_b

print(f'MSE:    {mse:.2f}')
print(f'grad_w: {grad_w:.4f}')
print(f'w_new:  {w_new:.4f}')
print(f'b_new:  {b_new:.4f}')
```

## Exercise 2: Full gradient descent

Run gradient descent for 1000 steps and plot the MSE loss curve.

```python
w = 0.0
b = 0.0
lr = 1e-6
losses = []

for step in range(1000):
    y_hat  = w * sizes + b
    errors = prices - y_hat
    mse    = np.mean(errors ** 2)
    losses.append(mse)
    grad_w = (-2 / n) * np.sum(sizes * errors)
    grad_b = (-2 / n) * np.sum(errors)
    w -= lr * grad_w
    b -= lr * grad_b

print(f'Final w: {w:.2f}, b: {b:.2f}')

plt.figure(figsize=(7, 4))
plt.plot(losses)
plt.title('MSE loss during gradient descent')
plt.xlabel('Step')
plt.ylabel('MSE')
plt.yscale('log')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## Exercise 3: Compare to scikit-learn

Fit the same data with `sklearn.linear_model.LinearRegression` and compare coefficients.

```python
from sklearn.linear_model import LinearRegression

X_sk = sizes.reshape(-1, 1)  # sklearn needs 2D input
model = LinearRegression()
model.fit(X_sk, prices)

print(f'sklearn  w: {model.coef_[0]:.2f},   b: {model.intercept_:.2f}')
print(f'from scratch w: {w:.2f},  b: {b:.2f}')
print('\nPredictions comparison:')
for sz, p_true in zip(sizes, prices):
    p_ours  = w * sz + b
    p_sklearn = model.predict([[sz]])[0]
    print(f'  size={sz:.0f}: true={p_true:.0f}, ours={p_ours:.0f}, sklearn={p_sklearn:.0f}')
```
