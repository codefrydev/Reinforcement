## Exercise 1 — Entropy Function

Implement H(y) = -sum(p_c * log2(p_c)) for a binary label array.

```python
def entropy(y):
    n = len(y)
    if n == 0:
        return 0.0
    counts = np.bincount(y)
    # TODO: compute probs and return entropy
    probs = counts[counts > 0] / n
    return None  # -np.sum(probs * np.log2(probs))

print(f'H([0,0,1,1]) = {entropy(np.array([0,0,1,1]))}')  # expected: 1.0
print(f'H([0,0,0])   = {entropy(np.array([0,0,0]))}')    # expected: 0.0
print(f'H([0,0,0,1,1,1]) = {entropy(np.array([0,0,0,1,1,1]))}')  # expected: 1.0
```

## Exercise 2 — Information Gain

Compute information gain for splitting x=[1..6], labels=[0,0,0,1,1,1] at x=3.5.

```python
labels = np.array([0, 0, 0, 1, 1, 1])
left   = labels[:3]  # x <= 3: [0, 0, 0]
right  = labels[3:]  # x >  3: [1, 1, 1]

H_parent   = entropy(labels)
H_left     = entropy(left)
H_right    = entropy(right)
n = len(labels)
# TODO: weighted entropy and information gain
weighted_H = (len(left)/n) * H_left + (len(right)/n) * H_right
info_gain  = None  # H_parent - weighted_H

print(f'H(parent) = {H_parent:.4f}')
print(f'H(left) = {H_left:.4f}, H(right) = {H_right:.4f}')
print(f'Information gain = {info_gain}')
# expected: 1.0 bit (perfect split)
```

## Exercise 3 — Imperfect Split

Try a different label arrangement and see how gain decreases.

```python
# labels not cleanly split at 3.5
labels2 = np.array([0, 0, 1, 0, 1, 1])
left2, right2 = labels2[:3], labels2[3:]

H2 = entropy(labels2)
wH2 = (len(left2)/len(labels2))*entropy(left2) + (len(right2)/len(labels2))*entropy(right2)
gain2 = H2 - wH2
print(f'Imperfect split gain = {gain2:.4f}')
print(f'Compare to perfect split gain = 1.0')
# expected: gain < 1.0 since split is not clean
```
