# Datasets and Features

In this notebook you will practice loading data with pandas, splitting a DataFrame into features X and labels y, and exploring datasets with standard inspection methods.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Exercise 1: Create and inspect a dataset

Build a small health dataset, inspect it with `.head()`, `.describe()`, and `.shape`.

```python
data = {
    'height': [165, 178, 155, 190, 172],
    'weight': [60,  80,  55,  95,  70],
    'age':    [25,  32,  28,  45,  38],
    'healthy': [1,   1,   1,   0,   1]
}
df = pd.DataFrame(data)

print('Shape:', df.shape)
print()
print(df.head())
print()
print(df.describe())
```

## Exercise 2: Split into X and y

Split the DataFrame into a feature matrix X (height, weight, age) and label vector y (healthy).

```python
# Split
X = df[['height', 'weight', 'age']]   # feature matrix
y = df['healthy']                      # label vector

print('X shape:', X.shape)             # expected: (5, 3)
print('y shape:', y.shape)             # expected: (5,)
print('X columns:', list(X.columns))
print()
print('Label distribution:')
print(y.value_counts())
```

## Exercise 3: Handle missing values

Add a missing value to the dataset, detect it, and fill it with the column mean.

```python
import numpy as np

df_missing = df.copy()
df_missing.loc[2, 'weight'] = np.nan   # introduce a missing value

print('Missing values per column:')
print(df_missing.isnull().sum())
print()

# Fill missing values with column mean
df_filled = df_missing.fillna(df_missing.mean(numeric_only=True))

print('After filling with mean:')
print(df_filled[['weight']])
```

## Exercise 4: Convert to NumPy for ML

Convert X and y to NumPy arrays — the format expected by most ML functions.

```python
X_np = X.values   # or X.to_numpy()
y_np = y.values

print('X_np dtype:', X_np.dtype)
print('X_np:\n', X_np)
print()
print('y_np:', y_np)
```
