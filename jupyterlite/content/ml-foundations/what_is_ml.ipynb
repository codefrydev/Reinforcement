## Exercise 2: Rule-based classifier (and its limits)

Write a simple if/else spam detector using keyword matching. Then test it on emails it hasn't seen — observe where rules fail.

```python
def rule_based_spam(email_text):
    """Return True if email looks like spam, False otherwise."""
    keywords = ['free', 'prize', 'winner', 'claim', 'offer', 'selected']
    email_lower = email_text.lower()
    return any(kw in email_lower for kw in keywords)

test_emails = [
    ('Win a FREE iPhone now!!!', True),
    ('Meeting at 3pm tomorrow', False),
    ('Claim your prize — limited time', True),
    ('Can you review my pull request?', False),
    ('You have been selected for exclusive offer', True),
    ('Congratulations on your achievement today', True),   # new — tricky
    ('Get rich quick with our secret system', True),       # new — tricky
]

correct_count = 0
for text, label in test_emails:
    pred = rule_based_spam(text)
    mark = 'OK  ' if pred == label else 'MISS'
    if pred == label:
        correct_count += 1
    print(f'{mark}: "{text[:50]}" -> pred={pred}, actual={label}')
print(f'\nAccuracy: {correct_count}/{len(test_emails)}')
```

## Exercise 3: Visualise ML problem type distribution

Plot a bar chart showing roughly how often each type of ML problem appears in industry.

```python
problem_types = ['Supervised', 'Unsupervised', 'Reinforcement']
counts = [1000, 300, 100]
colors = ['#4C72B0', '#DD8452', '#55A868']

fig, ax = plt.subplots(figsize=(7, 4))
bars = ax.bar(problem_types, counts, color=colors, edgecolor='white', linewidth=0.5)
ax.set_title('ML problem types in practice (approximate)', fontsize=13, fontweight='bold')
ax.set_ylabel('Approximate count in industry')
ax.set_xlabel('Problem type')

for bar, count in zip(bars, counts):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 10,
            str(count), ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()
```
