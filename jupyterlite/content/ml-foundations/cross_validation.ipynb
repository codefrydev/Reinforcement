## Exercise 2 — 5-Fold CV from Scratch

Implement cross-validation on a 20-sample synthetic dataset.

```python
np.random.seed(42)
X = np.random.randn(20, 2)
y = (X[:, 0] + X[:, 1] > 0).astype(int)

K = 5
fold_size = len(X) // K
scores = []

for fold in range(K):
    test_idx  = list(range(fold * fold_size, (fold + 1) * fold_size))
    # TODO: train_idx = complement of test_idx
    train_idx = [i for i in range(len(X)) if i not in test_idx]

    X_train, X_test = X[train_idx], X[test_idx]
    y_train, y_test = y[train_idx], y[test_idx]

    # majority class classifier
    majority = int(np.mean(y_train) >= 0.5)
    preds = np.full(len(y_test), majority)
    acc = np.mean(preds == y_test)
    scores.append(acc)

print(f'Fold scores: {[round(s, 3) for s in scores]}')
print(f'Mean CV: {np.mean(scores):.3f} +/- {np.std(scores):.3f}')
```

## Exercise 3 — Overfitting vs Underfitting

Compare degree-1 (underfit) and degree-5 (overfit) polynomial on a sine curve.

```python
np.random.seed(0)
x = np.linspace(0, 1, 20)
y = np.sin(2 * np.pi * x) + 0.1 * np.random.randn(20)

# split: first 15 train, last 5 test
x_train, y_train = x[:15], y[:15]
x_test,  y_test  = x[15:], y[15:]

for deg in [1, 5]:
    coeffs = np.polyfit(x_train, y_train, deg)
    train_err = np.mean((np.polyval(coeffs, x_train) - y_train)**2)
    test_err  = np.mean((np.polyval(coeffs, x_test)  - y_test )**2)
    print(f'Degree {deg}: train_MSE={train_err:.4f}, test_MSE={test_err:.4f}')
# expected: degree-5 has lower train error but higher test error (overfitting)
```
