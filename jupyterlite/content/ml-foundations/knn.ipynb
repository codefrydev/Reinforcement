## Exercise 3 — Effect of K

```python
test_point2 = np.array([2, 2], dtype=float)
distances2 = np.sqrt(np.sum((X_train - test_point2)**2, axis=1))
sorted_idx2 = np.argsort(distances2)

for K in [1, 3, 5, 7]:
    labels = y_train[sorted_idx2[:K]]
    pred = int(np.mean(labels) >= 0.5)
    print(f'K={K}: labels={labels}, prediction={pred}')
# observe how the boundary shifts with K
```
