# Logistic Regression

In this notebook you will implement logistic regression from scratch: sigmoid + cross-entropy loss + gradient descent. You will also compare your implementation to scikit-learn.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Exercise 1: Cross-entropy loss by hand

Compute cross-entropy for good and bad predictions to understand what the loss measures.

```python
def cross_entropy(y_true, y_pred):
    eps = 1e-9
    return -np.mean(y_true * np.log(y_pred + eps) + (1 - y_true) * np.log(1 - y_pred + eps))

y_true = np.array([1., 0., 1.])

# Good predictions
y_good = np.array([0.9, 0.1, 0.8])
print(f'Good predictions loss: {cross_entropy(y_true, y_good):.4f}')

# Random (50/50) predictions
y_random = np.array([0.5, 0.5, 0.5])
print(f'Random predictions loss: {cross_entropy(y_true, y_random):.4f}')

# Terrible predictions
y_bad = np.array([0.1, 0.9, 0.1])
print(f'Bad predictions loss:  {cross_entropy(y_true, y_bad):.4f}')
```

## Exercise 2: Train logistic regression from scratch

```python
np.random.seed(42)

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Toy dataset: 3 features, 5 samples
X = np.array([
    [1.0, 2.0, 0.5],
    [2.0, 1.0, 1.5],
    [0.5, 3.0, 0.2],
    [3.0, 0.5, 2.0],
    [1.5, 1.5, 1.0],
])
y = np.array([1., 1., 0., 1., 0.])

w  = np.zeros(3)
b  = 0.0
lr = 0.1
n  = len(y)
losses = []

for step in range(200):
    z    = X @ w + b
    p    = sigmoid(z)
    loss = cross_entropy(y, p)
    losses.append(loss)
    grad_w = (1/n) * X.T @ (p - y)
    grad_b = (1/n) * np.sum(p - y)
    w -= lr * grad_w
    b -= lr * grad_b
    if step % 40 == 0:
        print(f'step {step:3d}: loss={loss:.4f}')

print(f'\nFinal w: {w.round(3)}')
print(f'Final b: {b:.3f}')
```

## Exercise 3: Plot the training loss curve

```python
plt.figure(figsize=(7, 4))
plt.plot(losses, color='#4C72B0')
plt.title('Cross-entropy loss during logistic regression training')
plt.xlabel('Step')
plt.ylabel('Cross-entropy loss')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## Exercise 4: Compare to scikit-learn

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(C=1e6, max_iter=1000)   # C=1e6 = minimal regularisation
model.fit(X, y)

print('sklearn coef_:     ', model.coef_[0].round(3))
print(f'sklearn intercept_: {model.intercept_[0]:.3f}')
print()
print('our w:             ', w.round(3))
print('our b:             ', round(b, 3))
```
