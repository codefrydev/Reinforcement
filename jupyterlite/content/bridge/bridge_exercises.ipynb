    G = 0.0

import random

def random_walk(n_episodes=1000, seed=0):
    random.seed(seed)
    wins = 0
    for _ in range(n_episodes):
        pos = 2
        while pos not in (0, 4):
            pos += random.choice([-1, 1])
        if pos == 4:
            wins += 1
    return wins / n_episodes

def epsilon_greedy_action(Q_dict, state, n_actions=4, epsilon=0.1):
    if random.random() < epsilon:
        return random.randint(0, n_actions - 1)
    # Greedy: find max Q for this state

    return max(q_vals, key=q_vals.get)
