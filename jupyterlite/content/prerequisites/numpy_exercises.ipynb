## Exercise 4 — Argmax greedy policy
Given Q values for 4 actions, find the greedy action (argmax).

```python
Q_s = np.array([0.1, 0.5, 0.3, -0.2])
# TODO: greedy_action = np.argmax(Q_s)
greedy_action = None
print(f"Greedy action: {greedy_action}")
```

```python
Q_s = np.array([0.1, 0.5, 0.3, -0.2])
greedy_action = np.argmax(Q_s)
print(f"Greedy action: {greedy_action} (Q={Q_s[greedy_action]})")
```
