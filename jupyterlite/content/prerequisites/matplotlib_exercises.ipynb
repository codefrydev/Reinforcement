## Exercise 2 — Bandit arm comparison
Plot a bar chart comparing sample means of 5 arms: [0.2, 0.5, 0.1, 0.8, 0.4].

```python
arm_means = [0.2, 0.5, 0.1, 0.8, 0.4]
# TODO: bar chart with x labels 'Arm 1'...'Arm 5'
plt.figure(figsize=(5,3))
# your code
plt.show()
```

```python
arm_means = [0.2, 0.5, 0.1, 0.8, 0.4]
labels = [f'Arm {i+1}' for i in range(5)]
plt.figure(figsize=(5,3))
plt.bar(labels, arm_means, color='steelblue')
plt.ylabel('Sample Mean')
plt.title('Bandit Arm Estimates')
plt.tight_layout()
plt.show()
```

## Exercise 3 — Value function heatmap
Visualize a 3×3 value grid as a heatmap.

```python
V = np.array([[0.0, 0.3, 0.6], [0.2, 0.5, 0.8], [0.4, 0.7, 1.0]])
# TODO: plt.imshow(V), add colorbar, title 'Value Function'
plt.figure(figsize=(4,3))
# your code
plt.show()
```

```python
V = np.array([[0.0, 0.3, 0.6], [0.2, 0.5, 0.8], [0.4, 0.7, 1.0]])
plt.figure(figsize=(4,3))
plt.imshow(V, cmap='Blues')
plt.colorbar(label='V(s)')
plt.title('Value Function')
plt.tight_layout()
plt.show()
```
