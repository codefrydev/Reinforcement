## Exercise 2 — Linear value function

Linear FA: V(s) = w·φ(s). Given w=[0.5, -0.3, 0.8] and φ(s)=[1, 2, 0], compute V(s).

```python
w = np.array([0.5, -0.3, 0.8])
phi_s = np.array([1, 2, 0])
# TODO: V_s = np.dot(w, phi_s)
V_s = None
print(f"V(s) = {V_s}")
```

```python
w = np.array([0.5, -0.3, 0.8])
phi_s = np.array([1, 2, 0])
V_s = np.dot(w, phi_s)   # 0.5*1 + (-0.3)*2 + 0.8*0 = -0.1
print(f"V(s) = {V_s}")
```

## Exercise 3 — Gradient step

Gradient descent update: w ← w - α * ∇L. Given w=[1,2], gradient=[0.4, -0.6], α=0.1, compute new w.

```python
w = np.array([1.0, 2.0])
grad = np.array([0.4, -0.6])
alpha = 0.1
# TODO: w_new = w - alpha * grad
w_new = None
print(f"w_new = {w_new}")
```

```python
w = np.array([1.0, 2.0])
grad = np.array([0.4, -0.6])
alpha = 0.1
w_new = w - alpha * grad   # [0.96, 2.06]
print(f"w_new = {w_new}")
```
