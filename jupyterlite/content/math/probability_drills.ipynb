## Exercise 1 — Sample mean and variance

An arm has 4 samples: [1.2, 0.8, 1.0, 1.4]. Compute the sample mean and unbiased sample variance.

```python
x = [1.2, 0.8, 1.0, 1.4]
# TODO: compute mean = sum(x)/len(x)
# TODO: compute unbiased variance = sum((xi - mean)**2 for xi in x) / (len(x)-1)
mean = None
var = None
print(f"mean={mean}, var={var}")
```

```python
x = [1.2, 0.8, 1.0, 1.4]
mean = sum(x) / len(x)
var = sum((xi - mean)**2 for xi in x) / (len(x) - 1)
print(f"mean={mean:.2f}, var={var:.4f}")   # mean=1.10, var=0.0667
```

## Exercise 3 — Law of large numbers simulation

Simulate pulling a Bernoulli(0.4) arm 1000 times. Plot running sample mean.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
n = 1000
p_true = 0.4
samples = np.random.binomial(1, p_true, n)

# TODO: compute running mean at each step
# running_mean[i] = mean of samples[0:i+1]
running_mean = np.cumsum(samples) / np.arange(1, n+1)

plt.figure(figsize=(8, 3))
plt.plot(running_mean, label='Sample mean')
plt.axhline(p_true, color='red', linestyle='--', label=f'True mean ({p_true})')
plt.xlabel('Samples')
plt.ylabel('Mean')
plt.title('Sample mean converges to E[X] (Law of Large Numbers)')
plt.legend()
plt.tight_layout()
plt.show()
```
