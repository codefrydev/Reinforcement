## Exercise 3 — Policy gradient sign

Policy gradient: if log π(a|s) is high and reward is positive, gradient is positive (increase this action). Verify by computing log probability and the gradient signal.

```python
import numpy as np

# Softmax policy over 3 actions
logits = np.array([1.0, 2.0, 0.5])
probs = np.exp(logits) / np.sum(np.exp(logits))
print("Action probabilities:", np.round(probs, 3))

action_taken = 1   # action index
reward = 1.0

# Policy gradient signal = reward * (1 - prob_of_action_taken) for chosen action
# (simplified for one-step)
grad_signal = reward * (1 - probs[action_taken])
print(f"Gradient signal for action {action_taken}: {grad_signal:.4f}")
print("Positive signal means: increase probability of this action")
```
