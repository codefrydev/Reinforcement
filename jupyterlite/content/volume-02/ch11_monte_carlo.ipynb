```python
def random_walk_episode():
    state = 3
    trajectory = []
    while state not in (0, 5):
        next_state = state + (1 if np.random.random() < 0.5 else -1)
        reward = 1 if next_state == 5 else 0
        trajectory.append((state, reward))
        state = next_state
    return trajectory

for _ in range(3):
    t = random_walk_episode()
    print(f"Last step: {t[-1]}, length={len(t)}")
```

```python
def mc_prediction(n_episodes=5000):
    returns = {s: [] for s in range(1, 5)}
    for _ in range(n_episodes):
        episode = random_walk_episode()
        states, rewards = zip(*episode) if episode else ([], [])
        states, rewards = list(states), list(rewards)
        visited = set()
        for i, s in enumerate(states):
            if s in returns and s not in visited:
                visited.add(s)
                G = sum(rewards[i:])
                returns[s].append(G)
    return {s: np.mean(returns[s]) if returns[s] else 0.0 for s in range(1, 5)}

V = mc_prediction(5000)
for s in range(1, 5):
    print(f"V({s}) = {V[s]:.3f}  (true = {s/5:.1f})")

# Plot
states = list(range(1, 5))
mc_vals = [V[s] for s in states]
true_vals = [s/5 for s in states]
plt.figure(figsize=(5,3))
plt.plot(states, mc_vals, marker='o', label='MC estimate')
plt.plot(states, true_vals, linestyle='--', label='True V(s)')
plt.xlabel('State')
plt.ylabel('V(s)')
plt.title('MC Prediction: Random Walk')
plt.legend()
plt.tight_layout()
plt.show()
```
