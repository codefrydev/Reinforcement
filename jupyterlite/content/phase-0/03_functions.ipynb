## Exercise 2 — discounted return

```python
# TODO: define discounted_return(rewards, gamma) that returns sum(gamma**t * r for t, r in enumerate(rewards))
def discounted_return(rewards, gamma):
    pass

# Test: should print 0.81
print(discounted_return([0, 0, 1], 0.9))
```

```python
def discounted_return(rewards, gamma):
    return sum(gamma**t * r for t, r in enumerate(rewards))

print(discounted_return([0, 0, 1], 0.9))   # 0.81
print(discounted_return([1, 0, 0], 0.9))   # 1.0
print(discounted_return([1, 1, 1], 0.9))   # 2.71
```
