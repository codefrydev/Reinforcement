```python
def step(state, action):
    row, col = state
    moves = {0: (-1, 0), 1: (1, 0), 2: (0, -1), 3: (0, 1)}
    dr, dc = moves[action]
    nr, nc = row + dr, col + dc
    if not (0 <= nr <= 2 and 0 <= nc <= 2):
        return (row, col), -1, False
    if (nr, nc) == (2, 2):
        return (2, 2), 1, True
    return (nr, nc), 0, False

print(step((0, 0), 3))   # ((0, 1), 0, False)
print(step((2, 1), 3))   # ((2, 2), 1, True)
print(step((0, 0), 0))   # ((0, 0), -1, False)
```
