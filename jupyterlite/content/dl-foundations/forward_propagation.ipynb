## Exercise 1: Single-Sample Forward Pass

Implement z1, h1, z2, ŷ for a 2-layer MLP and print all intermediate shapes.

```python
import numpy as np
np.random.seed(42)

x  = np.array([0.5, -0.3, 0.8])
W1 = np.random.randn(4, 3)
b1 = np.random.randn(4)
W2 = np.random.randn(2, 4)
b2 = np.random.randn(2)

def softmax(z):
    e = np.exp(z - np.max(z))
    return e / np.sum(e)

# Forward pass
z1    = W1 @ x + b1
h1    = np.maximum(0, z1)
z2    = W2 @ h1 + b2
y_hat = softmax(z2)

print('Shapes through the network:')
print(f'  x     : {x.shape}    → {x.round(2)}')
print(f'  z1    : {z1.shape}  → {z1.round(3)}')
print(f'  h1    : {h1.shape}  → {h1.round(3)}  (ReLU zeros out negatives)')
print(f'  z2    : {z2.shape}  → {z2.round(3)}')
print(f'  y_hat : {y_hat.shape}  → {y_hat.round(4)}  (sum={y_hat.sum():.4f})')
```

## Exercise 2: Batch Forward Pass

Process a batch of 5 inputs at once using matrix operations.

```python
import numpy as np
np.random.seed(42)

X_batch = np.random.randn(5, 3)  # 5 samples, 3 features each
W1 = np.random.randn(4, 3)
b1 = np.random.randn(4)
W2 = np.random.randn(2, 4)
b2 = np.random.randn(2)

def softmax_batch(Z):
    E = np.exp(Z - Z.max(axis=1, keepdims=True))
    return E / E.sum(axis=1, keepdims=True)

# Batch forward pass: note transposed weight convention for batch
z1    = X_batch @ W1.T + b1       # (5, 3) × (3, 4) → (5, 4)
h1    = np.maximum(0, z1)          # (5, 4)
z2    = h1 @ W2.T + b2             # (5, 4) × (4, 2) → (5, 2)
y_hat = softmax_batch(z2)          # (5, 2)

print('Batch forward pass shapes:')
print(f'  X_batch : {X_batch.shape}')
print(f'  z1      : {z1.shape}')
print(f'  h1      : {h1.shape}')
print(f'  y_hat   : {y_hat.shape}')
print(f'  Row sums: {y_hat.sum(axis=1).round(4)}  ← all 1.0')
```

## Exercise 3: DQN-Style Network

Build a CartPole Q-network: state(8) → 64 → 64 → Q-values(2)

```python
import numpy as np
np.random.seed(0)

# CartPole-style state
state = np.array([0.02, -0.01, 0.04, 0.03, -0.02, 0.01, 0.0, 0.0])

# DQN network weights (Xavier initialization)
W1 = np.random.randn(64, 8)  * np.sqrt(2.0/8)
b1 = np.zeros(64)
W2 = np.random.randn(64, 64) * np.sqrt(2.0/64)
b2 = np.zeros(64)
W3 = np.random.randn(2, 64)  * np.sqrt(2.0/64)
b3 = np.zeros(2)

# Forward pass
h1 = np.maximum(0, W1 @ state + b1)
h2 = np.maximum(0, W2 @ h1 + b2)
q_values = W3 @ h2 + b3  # no activation: Q-values are unbounded

greedy_action = np.argmax(q_values)
actions = ['push left', 'push right']

print('DQN forward pass:')
print(f'  State:          {state}')
print(f'  Q(s, left):     {q_values[0]:.4f}')
print(f'  Q(s, right):    {q_values[1]:.4f}')
print(f'  Greedy action:  {greedy_action} ({actions[greedy_action]})')
```
