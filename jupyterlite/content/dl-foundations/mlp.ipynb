## Exercise 1: Parameter Counting

Count parameters in a 3-layer MLP: 4 → 8 → 4 → 2

```python
import numpy as np

layers = [4, 8, 4, 2]  # input=4, hidden1=8, hidden2=4, output=2

total = 0
for i in range(len(layers) - 1):
    n_in, n_out = layers[i], layers[i+1]
    params = n_in * n_out + n_out  # weights + biases
    print(f'Layer {i+1}: {n_in} → {n_out} | W: {n_out}×{n_in}={n_in*n_out}, b: {n_out} | subtotal: {params}')
    total += params

print(f'\nTotal parameters: {total}')
```

## Exercise 2: Solving XOR with a 2-Layer MLP

Use pre-designed weights to show XOR is solvable with two layers.

```python
import numpy as np

# Hand-designed weights that solve XOR
W1 = np.array([[1.0,  1.0],
               [-1.0, -1.0]])
b1 = np.array([-0.5, 1.5])
W2 = np.array([[1.0, -1.0]])
b2 = np.array([-0.5])

XOR_inputs  = np.array([[0,0],[0,1],[1,0],[1,1]], dtype=float)
XOR_outputs = np.array([0, 1, 1, 0])

print('XOR forward pass with 2-layer MLP:')
print('-' * 40)
all_correct = True
for x, y_true in zip(XOR_inputs, XOR_outputs):
    h = np.maximum(0, W1 @ x + b1)       # ReLU hidden layer
    out = W2 @ h + b2                     # linear output
    y_hat = 1 if out[0] > 0 else 0
    correct = '✓' if y_hat == y_true else '✗'
    print(f'  x={x.astype(int)}, h={h}, y_hat={y_hat}, y_true={y_true}  {correct}')
    if y_hat != y_true:
        all_correct = False

print(f'\nAll correct: {all_correct}')
```

## Exercise 3: Visualize XOR Decision Regions

Plot a fine grid of points colored by the MLP's output to see the non-linear boundary.

```python
import numpy as np
import matplotlib.pyplot as plt

# Grid
xx, yy = np.meshgrid(np.linspace(-0.5, 1.5, 200), np.linspace(-0.5, 1.5, 200))
grid = np.c_[xx.ravel(), yy.ravel()]

# Predict on grid
preds = []
for pt in grid:
    h = np.maximum(0, W1 @ pt + b1)
    out = W2 @ h + b2
    preds.append(1 if out[0] > 0 else 0)
preds = np.array(preds).reshape(xx.shape)

fig, ax = plt.subplots(figsize=(6, 6))
ax.contourf(xx, yy, preds, alpha=0.3, cmap='RdBu')

XOR_inputs = np.array([[0,0],[0,1],[1,0],[1,1]], dtype=float)
XOR_outputs = np.array([0, 1, 1, 0])
colors = ['blue' if y == 0 else 'red' for y in XOR_outputs]
ax.scatter(XOR_inputs[:,0], XOR_inputs[:,1], c=colors, s=200, zorder=5, edgecolors='black')

for pt, y in zip(XOR_inputs, XOR_outputs):
    ax.annotate(str(y), pt, textcoords='offset points', xytext=(8, 8), fontsize=12, fontweight='bold')

ax.set_title('XOR: MLP Decision Regions (non-linear boundary)')
ax.set_xlabel('x₁')
ax.set_ylabel('x₂')
plt.tight_layout()
plt.show()
```
