## Exercise 1: Implement and Plot All Activations

Implement ReLU, sigmoid, tanh, and plot them together.

```python
import numpy as np
import matplotlib.pyplot as plt

z = np.linspace(-4, 4, 300)

relu    = np.maximum(0, z)
sigmoid = 1 / (1 + np.exp(-z))
tanh    = np.tanh(z)

fig, axes = plt.subplots(1, 3, figsize=(14, 4))

for ax, fn, name, color in zip(axes,
        [relu, sigmoid, tanh],
        ['ReLU', 'Sigmoid', 'Tanh'],
        ['steelblue', 'darkorange', 'green']):
    ax.plot(z, fn, color=color, linewidth=2)
    ax.axhline(0, color='gray', linewidth=0.5)
    ax.axvline(0, color='gray', linewidth=0.5)
    ax.set_title(name, fontsize=13)
    ax.set_xlabel('z')
    ax.set_ylabel('f(z)')
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## Exercise 2: Softmax and Numerical Stability

Implement softmax with the max-subtraction trick and verify it handles large inputs.

```python
import numpy as np

def softmax_unstable(z):
    exp_z = np.exp(z)
    return exp_z / np.sum(exp_z)

def softmax_stable(z):
    exp_z = np.exp(z - np.max(z))  # subtract max for numerical stability
    return exp_z / np.sum(exp_z)

# Normal inputs — both work
z_normal = np.array([1.0, 2.0, 3.0])
print('Normal inputs:')
print(f'  Unstable: {softmax_unstable(z_normal).round(4)}')
print(f'  Stable:   {softmax_stable(z_normal).round(4)}')

# Large inputs — unstable version overflows
z_large = np.array([1000.0, 1001.0, 1002.0])
print('\nLarge inputs (would overflow without stability fix):')
print(f'  Unstable: {softmax_unstable(z_large)}')
print(f'  Stable:   {softmax_stable(z_large).round(4)}')
```

## Exercise 3: Dead ReLU Demo

Show that when all inputs to a ReLU neuron are negative, its gradient is always 0.

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate a neuron with always-negative pre-activations
np.random.seed(7)
z_dead = np.random.randn(100) - 3.0  # shift mean to -3 so all are negative

relu_output   = np.maximum(0, z_dead)
relu_gradient = (z_dead > 0).astype(float)

print(f'Pre-activations: min={z_dead.min():.2f}, max={z_dead.max():.2f}')
print(f'ReLU outputs:    all zero? {np.all(relu_output == 0)}')
print(f'ReLU gradients:  all zero? {np.all(relu_gradient == 0)}')
print('→ Dead neuron: no gradient flows, weights never update!')

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))
ax1.hist(z_dead, bins=20, color='steelblue', edgecolor='white')
ax1.set_title('Pre-activations (all negative)')
ax1.set_xlabel('z')
ax2.hist(relu_output, bins=20, color='coral', edgecolor='white')
ax2.set_title('ReLU output (all zero)')
ax2.set_xlabel('ReLU(z)')
plt.tight_layout()
plt.show()
```
