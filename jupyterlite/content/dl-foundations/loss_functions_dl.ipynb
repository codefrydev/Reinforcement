## Exercise 1: Implement MSE and Binary Cross-Entropy

```python
import numpy as np

def mse(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

def binary_cross_entropy(y_true, y_pred, eps=1e-7):
    p = np.clip(y_pred, eps, 1 - eps)
    return np.mean(-(y_true * np.log(p) + (1 - y_true) * np.log(1 - p)))

# BCE test
y_true = np.array([1, 0, 1, 1, 0])
good_preds = np.array([0.9, 0.1, 0.8, 0.9, 0.1])  # confident and correct
bad_preds  = np.array([0.6, 0.5, 0.5, 0.6, 0.4])  # uncertain

print('Binary Cross-Entropy:')
print(f'  Good predictions: {binary_cross_entropy(y_true, good_preds):.4f}')
print(f'  Bad  predictions: {binary_cross_entropy(y_true, bad_preds):.4f}')
print(f'  Lower is better: {binary_cross_entropy(y_true, good_preds) < binary_cross_entropy(y_true, bad_preds)}')

# MSE test
y_reg_true = np.array([2.5, 3.0, 1.5, 4.0])
y_reg_pred = np.array([2.3, 3.2, 1.8, 3.7])
print(f'\nMSE: {mse(y_reg_true, y_reg_pred):.4f}')
```

## Exercise 2: Visualize the MSE Loss Landscape

For a trivial 1-parameter model `ŷ = w * x`, vary `w` and plot MSE vs `w`.

```python
import numpy as np
import matplotlib.pyplot as plt

# Dataset: y = 2x (true weight = 2)
np.random.seed(42)
x_data = np.random.randn(20)
y_data = 2.0 * x_data + np.random.randn(20) * 0.3

# Sweep w and compute MSE
w_range = np.linspace(-1, 5, 200)
losses  = [np.mean((w * x_data - y_data)**2) for w in w_range]

plt.figure(figsize=(8, 5))
plt.plot(w_range, losses, 'steelblue', linewidth=2)
plt.axvline(2.0, color='red', linestyle='--', label='True w = 2.0')
plt.xlabel('w (model parameter)')
plt.ylabel('MSE Loss')
plt.title('Loss Landscape: MSE vs Parameter Value')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
print(f'Minimum loss at w ≈ {w_range[np.argmin(losses)]:.2f}  (true: 2.0)')
```

## Exercise 3: DQN TD Error Loss

Implement the DQN loss: `L = (r + γ * max Q(s',a') - Q(s,a))²`

```python
import numpy as np

# Simulate a batch of 4 DQN transitions
np.random.seed(1)

gamma = 0.99
rewards        = np.array([1.0,  0.0,  0.0, -1.0])  # r
dones          = np.array([False, False, False, True])  # episode ends

# Simulated Q-values from the main and target network
q_current = np.array([0.5,  0.3,  0.7, 0.2])  # Q(s,a) for taken action
q_next_max = np.array([0.8, 0.6,  0.9, 0.0])   # max_a' Q(s',a') from target net

# Bellman targets: if done, target = r; else target = r + gamma * max Q(s',a')
targets = rewards + gamma * q_next_max * (~dones).astype(float)

# TD errors and MSE loss
td_errors = targets - q_current
dqn_loss  = np.mean(td_errors ** 2)

print('DQN Transition Batch:')
print(f'  rewards:    {rewards}')
print(f'  targets:    {targets.round(3)}')
print(f'  q_current:  {q_current}')
print(f'  TD errors:  {td_errors.round(3)}')
print(f'  DQN loss:   {dqn_loss:.4f}')
```
