## Exercise 1: Single Neuron

Implement a neuron: `output = activation(w · x + b)`

Given: `x = [0.5, 0.3, 0.8]`, `w = [0.2, -0.5, 0.4]`, `b = 0.1`

```python
import numpy as np

x = np.array([0.5, 0.3, 0.8])
w = np.array([0.2, -0.5, 0.4])
b = 0.1

# Compute pre-activation
z = np.dot(w, x) + b

# Activations
step    = 1 if z > 0 else 0
sigmoid = 1 / (1 + np.exp(-z))
relu    = max(0, z)

print(f'z       = {z:.4f}')
print(f'step    = {step}')
print(f'sigmoid = {sigmoid:.4f}')
print(f'relu    = {relu:.4f}')
```

## Exercise 2: Visualize Activation Functions

Plot sigmoid, tanh, and ReLU over `z ∈ [-4, 4]` on the same axes.

```python
import numpy as np
import matplotlib.pyplot as plt

z = np.linspace(-4, 4, 200)

sigmoid = 1 / (1 + np.exp(-z))
tanh    = np.tanh(z)
relu    = np.maximum(0, z)

plt.figure(figsize=(8, 5))
plt.plot(z, sigmoid, label='Sigmoid', linewidth=2)
plt.plot(z, tanh,    label='Tanh',    linewidth=2)
plt.plot(z, relu,    label='ReLU',    linewidth=2)
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.xlabel('z (pre-activation)')
plt.ylabel('f(z)')
plt.title('Activation Functions')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## Exercise 3: Weight Sensitivity

Vary each weight one at a time and observe how z changes.
This builds intuition for what weights 'do' in a neural network.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0.5, 0.3, 0.8])
b = 0.1

# Sweep w[0] from -3 to 3, keeping w[1]=−0.5, w[2]=0.4 fixed
w0_range = np.linspace(-3, 3, 100)
z_values = [np.dot(np.array([w0, -0.5, 0.4]), x) + b for w0 in w0_range]
sigmoid_values = [1 / (1 + np.exp(-z)) for z in z_values]

plt.figure(figsize=(8, 4))
plt.plot(w0_range, sigmoid_values, 'b-', linewidth=2)
plt.xlabel('w[0] (weight for x[0]=0.5)')
plt.ylabel('sigmoid output')
plt.title('Neuron output as w[0] varies')
plt.axhline(0.5, color='red', linestyle='--', alpha=0.5, label='threshold (0.5)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
