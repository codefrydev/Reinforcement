## Exercise 1: Perceptron on AND Gate

Implement the perceptron learning rule: `w += α(y − ŷ)x`

```python
import numpy as np

X = np.array([[0,0],[0,1],[1,0],[1,1]], dtype=float)
y_and = np.array([0, 0, 0, 1])

def perceptron_train(X, y, epochs=20, alpha=1.0):
    w = np.zeros(X.shape[1])
    b = 0.0
    history = []
    for epoch in range(epochs):
        errors = 0
        for i in range(len(X)):
            y_hat = 1 if np.dot(w, X[i]) + b > 0 else 0
            if y_hat != y[i]:
                w += alpha * (y[i] - y_hat) * X[i]
                b += alpha * (y[i] - y_hat)
                errors += 1
        preds = np.array([1 if np.dot(w, X[i]) + b > 0 else 0 for i in range(len(X))])
        acc = np.mean(preds == y)
        history.append(acc)
        if errors == 0:
            print(f'Converged at epoch {epoch+1}')
            break
    return w, b, history

w, b, history = perceptron_train(X, y_and)
print(f'Final weights: w={w}, b={b:.2f}')
print(f'Final accuracy: {history[-1]:.0%}')
```

## Exercise 2: Visualize Decision Boundary

Plot the AND gate points and the learned decision boundary `w·x + b = 0`.

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(6, 6))

# Plot points
X = np.array([[0,0],[0,1],[1,0],[1,1]], dtype=float)
y_and = np.array([0, 0, 0, 1])
colors = ['blue' if yi == 0 else 'red' for yi in y_and]
ax.scatter(X[:,0], X[:,1], c=colors, s=200, zorder=5)

# Decision boundary: w[0]*x1 + w[1]*x2 + b = 0  =>  x2 = -(w[0]*x1 + b) / w[1]
if w[1] != 0:
    x1_range = np.array([-0.5, 1.5])
    x2_boundary = -(w[0] * x1_range + b) / w[1]
    ax.plot(x1_range, x2_boundary, 'g-', linewidth=2, label='Decision boundary')

ax.set_xlim(-0.3, 1.3)
ax.set_ylim(-0.3, 1.3)
ax.set_xlabel('x₁')
ax.set_ylabel('x₂')
ax.set_title('AND Gate: Perceptron Decision Boundary')
ax.legend()
ax.grid(True, alpha=0.3)

# Labels
for i, (xi, yi) in enumerate(zip(X, y_and)):
    ax.annotate(f'({int(xi[0])},{int(xi[1])})→{yi}', xi, textcoords='offset points',
                xytext=(8, 8), fontsize=10)
plt.tight_layout()
plt.show()
```

## Exercise 3: XOR — The Limitation

Try training the same perceptron on XOR. Show that it never converges.

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([[0,0],[0,1],[1,0],[1,1]], dtype=float)
y_xor = np.array([0, 1, 1, 0])

_, _, history_xor = perceptron_train(X, y_xor, epochs=100)

plt.figure(figsize=(8, 4))
plt.plot(history_xor, 'r-o', markersize=4)
plt.axhline(1.0, color='green', linestyle='--', label='100% accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('XOR: Perceptron Never Converges')
plt.ylim(0, 1.1)
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
print(f'Best XOR accuracy: {max(history_xor):.0%} (never reaches 100%)')
```
